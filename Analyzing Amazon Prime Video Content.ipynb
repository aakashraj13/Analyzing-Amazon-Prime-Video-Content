### Variables Description 

The Amazon Prime Video dataset consists of two files: titles.csv and credits.csv, each containing multiple variables that describe the available content and its associated cast and crew.

Variables in titles.csv (Metadata of Movies & TV Shows)

- id – A unique identifier for each title from JustWatch.
- title – The name of the movie or TV show.
- type – Specifies whether the title is a TV Show or Movie.
- description – A brief summary of the title’s plot or content.
- release_year – The year the title was released.
- age_certification – The age rating assigned to the title (e.g., PG, R, etc.).
- runtime – Duration of the movie or the average length of a TV show episode (in minutes).
- genres – The category or genre(s) the title belongs to (e.g., Drama, Comedy, Action).
- production_countries – The country or countries involved in producing the title.
- seasons – Number of seasons (applicable only for TV shows).
- imdb_id – The corresponding IMDb identifier for the title.
- imdb_score – The IMDb rating for the title (out of 10).
- imdb_votes – The total number of votes the title has received on IMDb.
- tmdb_popularity – Popularity score from TMDB (The Movie Database).
- tmdb_score – The rating of the title on TMDB.

Variables in credits.csv (Cast & Crew Details)

- person_id – A unique identifier for each individual (actor or director).
- id – The JustWatch title ID, linking the person to a specific title in titles.csv.
- name – The name of the actor or director.
- character – The name of the character played by the actor (missing for directors).
- role – Specifies whether the individual is an ACTOR or DIRECTOR.

These variables collectively provide insights into content trends, regional availability, audience preferences, IMDb and TMDB ratings, and the influence of actors and directors in Amazon Prime Video’s content catalog.

## 3. ***Data Wrangling***

### Data Wrangling Code

```python
# Write your code to make your dataset analysis ready. 
df1.head(4)  

#Dropping duplicates from primary key ID 
df1 = df1.drop_duplicates(subset=['id'], keep='first') 

#handling missing values 
df1['age_certification'].fillna('Unknown', inplace=True) #Filling with "Unknown" or "Not Rated".
df1['seasons'].fillna(0, inplace=True) #Filling with 0 (assuming missing values belong to movies).
df1['imdb_score'].fillna(df1['imdb_score'].median(), inplace=True)
df1['tmdb_score'].fillna(df1['tmdb_score'].median(), inplace=True)
df1['description'].fillna('No description available', inplace=True)

#filling nan with N/A in character
df2['character'].fillna('N/A', inplace=True)

import ast # To safely convert string representations of lists
df1['genres'] = df1['genres'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)
# Explode the 'genres' column (each genre gets a separate row)
df_genres = df1.explode('genres')

# Remove any empty genre entries
df_genres = df_genres[df_genres['genres'].str.strip() != '']

# Convert string representation of list to an actual list (if needed)
df1['production_countries'] = df1['production_countries'].apply(lambda x: eval(x) if isinstance(x, str) else x) 

#merge to tables
df = df1.merge(df2, on='id', how='inner')
```

#### Chart - 2 Trends in Genre Distribution Over the Years

```python
# Count the number of titles per genre each year
genre_trends = df_genres.groupby(['release_year', 'genres']).size().reset_index(name='count')

# Select top 6 most common genres for better clarity
top_genres = genre_trends.groupby('genres')['count'].sum().sort_values(ascending=False).head(6).index
genre_trends_top = genre_trends[genre_trends['genres'].isin(top_genres)]

# Apply a rolling mean for smoothing (window size = 3 years)
genre_trends_top = genre_trends_top.copy()
genre_trends_top['count_smoothed'] = genre_trends_top.groupby('genres')['count'].transform(lambda x: x.rolling(3, min_periods=1).mean())

# Plot genre trends over time
plt.figure(figsize=(14, 7))
sns.lineplot(data=genre_trends_top, x='release_year', y='count_smoothed', hue='genres', 
             marker='o', linewidth=2.5, palette="Set2")

plt.xlabel("Release Year", fontsize=12)
plt.ylabel("Number of Titles ", fontsize=12)
plt.title("Trends in Genre Distribution Over the Years ", fontsize=14)
plt.legend(title="Genres")
plt.grid(True, linestyle="--", alpha=0.7)

plt.show()
```

##### 1. Why did you pick the specific chart?
A line chart was chosen because it effectively shows trends over time. Since we are analyzing how the popularity of different genres has changed across years, a line plot helps visualize increases, decreases, and fluctuations in genre frequency. It allows us to compare multiple genres at once and identify long-term patterns, making it ideal for this type of time-series data analysis.

##### 2. What is/are the insight(s) found from the chart?
The chart reveals key trends in genre popularity over time. Certain genres, like drama and comedy, have remained consistently popular, while others, such as thrillers and documentaries, have seen fluctuations. #### Chart - 11 Top 10 Highest-Rated Actors (Filtered by Minimum 3 Titles)

```python

# Filter only actors
df_actors = df[df["role"] == "ACTOR"]

# Calculate the average IMDb and TMDB scores + count of appearances
top_actors = df_actors.groupby("name").agg(
    imdb_score=('imdb_score', 'mean'),
    tmdb_score=('tmdb_score', 'mean'),
    count=('title', 'count')  # Number of titles an actor appeared in
).reset_index()

# Apply a minimum threshold (e.g., actors with at least 3 appearances)
top_actors = top_actors[top_actors['count'] >= 3]

# Sort by IMDb score first, then TMDB score, then by the number of appearances
top_actors = top_actors.sort_values(by=["imdb_score", "tmdb_score", "count"], ascending=[False, False, False])

# Get the top 10 highest-rated actors
top_actors = top_actors.head(10)

# Plot the top actors
plt.figure(figsize=(10, 5))
ax = sns.barplot(x='imdb_score', y='name', data=top_actors,hue = 'name', palette='hls')

# Add value labels inside the bars
for index, value in enumerate(top_actors["imdb_score"]):
    ax.text(value - 0.1, index, f"{value:.1f}", color="white", ha="right", va="center", fontweight="bold")

plt.xlabel("Average IMDb Score")
plt.ylabel("Actor")
plt.title("Top 10 Highest-Rated Actors (Filtered by Minimum 3 Titles)")
plt.xlim(0, 10)  # Set x-axis limit for better visibility
plt.show()
```
